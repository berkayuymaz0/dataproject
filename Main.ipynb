```python
df_continentconsumptioncopy = df_continentconsumption.copy()
df_continentconsumptioncopy = df_continentconsumptioncopy[df_continentconsumptioncopy["Year"] == 2020]
colors = ["khaki", "lightcoral", "lightslategray", "mediumseagreen",  "lightseagreen", "palevioletred", 'lightcoral', 'lightblue', 'pink']

df_continentconsumptioncopy_transpose = df_continentconsumptioncopy.T
df_continentconsumptioncopy_transpose = df_continentconsumptioncopy_transpose.sort_values(by=[30], ascending=False)
df_continentconsumptioncopy_transpose = df_continentconsumptioncopy_transpose.head(len(df_continentconsumptioncopy_transpose)-2)
df_continentconsumptioncopy = df_continentconsumptioncopy_transpose.T

total_values = df_continentconsumptioncopy.values.flatten().tolist()[1:]
counry_names = df_continentconsumptioncopy.columns.tolist()[1:]

plt.pie(total_values, labeldistance=1.1, labels=counry_names, autopct="%1.1f%%", 
        colors=colors, wedgeprops = { "linewidth" : 1.5, "edgecolor" : "white" })
plt.title("Total Energy Consumption around the World")
plt.show()
```

```python
#This figure shows Total Renewable Energy Production in Top 20 Countries
pd.options.mode.chained_assignment = None 

df_Country_Consumption_copy = df_Country_Consumption.copy()
df_renewable_power_generation_top20_copy = df_renewable_power_generation_top20.copy()

df_renewable_power_generation_top20_copy["Total Consumption(TWh)"] = None
df_renewable_power_generation_top20_copy["Country"][1] = "United States" 
df_renewable_power_generation_top20_copy["Country"][10] = "United Kingdom" 

df_Country_Consumption_copy= df_Country_Consumption_copy[df_Country_Consumption_copy["Year"] == 2020.0]

for rows in range(len(df_renewable_power_generation_top20_copy)):
    country = df_renewable_power_generation_top20_copy["Country"].tolist()[rows]
    total_consumption = df_Country_Consumption_copy[country].tolist()[0]*100
    df_renewable_power_generation_top20_copy["Total Consumption(TWh)"][rows] = total_consumption
    df_renewable_power_generation_top20_copy["Total (TWh)"][rows] = df_renewable_power_generation_top20_copy["Total (TWh)"][rows]*100

display(df_renewable_power_generation_top20_copy)

df_renewable_power_generation_top20_copy = df_renewable_power_generation_top20_copy.sort_values(by="Total Consumption(TWh)", ascending=False)


fig, ax1 = plt.subplots(figsize=(15, 10))
ax2 = ax1.twinx()
sn.barplot(data = df_renewable_power_generation_top20_copy, x='Country', y='Total Consumption(TWh)', ax=ax1)
ax1.set(ylim=(0, 350000))
ax1.tick_params(axis='x', labelrotation = 45)
sn.pointplot(data = df_renewable_power_generation_top20_copy, x='Country', y='Total (TWh)', ax=ax2)
ax2.set(ylim=(0, 350000))
ax2.tick_params(axis='x', labelrotation = 45)
plt.title("Total Renewable Energy Production in Top 20 Countries")
plt.show()
```
