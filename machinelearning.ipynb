```python
# Import numpy and pandas package 
import pandas as pd
import numpy as np 
# Data visualization
from matplotlib import pyplot as plot
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datas/Continent_Consumption_TWH.csv")
df
```

```python
df = df.copy()

df.plot(x="Year", y="World", title="World Energy Consumption",legend="brief", color="darkred")
plt.ylabel("TWh")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
                                                                                                              
```

```python
# with columns "Year" and "World"

# Extract the year and world data as numpy arrays
X = df['Year'].values.reshape(-1, 1)  # year values, reshaped to 2D array
y = df['World'].values  # world values

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Use the model to make predictions for the next year
next_year = model.predict(np.array([[2021]]))
print(f'Prediction for 2021: {next_year[0]:.2f}')
```

```python
df_capacity = pd.read_csv("datas/installed-global-renewable-energy-capacity-by-technology.csv")
df_capacity = df_capacity.drop(columns=["Code"])
df_capacity = df_capacity.iloc[:-23]
df_capacity = df_capacity.tail(21)
df_capacity = df_capacity.reset_index(drop=True)
df_capacity["Capacity"] = df_capacity["Capacity"].div(1000000)
df_capacity
```

```python
df2 = pd.read_csv("datas/renewable-energy-gen.csv")
df3 = df2.tail(32)
new_df = df3.copy()
new_df = new_df.iloc[:-1]
del new_df["Code"]
new_df = new_df.copy()
new_df = new_df.assign(Generated_Renewable_Energy=new_df["Solar Generation - TWh"] + new_df["Wind Generation - TWh"] + new_df["Hydro Generation - TWh"] + new_df["Geo Biomass Other - TWh"] )
new_df = new_df.drop(columns=["Solar Generation - TWh", "Wind Generation - TWh", "Hydro Generation - TWh", "Geo Biomass Other - TWh"])
new_df = new_df.tail(21)
new_df = new_df.reset_index(drop=True)
new_df = new_df = new_df.assign(Capacity=df_capacity["Capacity"])
new_df
```

```python
# with columns "Year" and "World"

# Extract the year and world data as numpy arrays
X2 = np.concatenate((new_df['Year'].values.reshape(-1, 1), new_df['Capacity'].values.reshape(-1, 1)), axis=1)
y2 = new_df['Generated_Renewable_Energy'].values  # world values

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X2, y2)

# Use the model to make predictions for the next year
# we assume capacity will be 3 by 2021
next_year2 = model.predict(np.array([[2021, 3]]))
print(f'Prediction for 2021: {next_year2[0]:.2f}')
```
